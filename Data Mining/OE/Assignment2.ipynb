```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
org_data = pd.read_excel('Data II.xlsx')
```

```python
org_data
```

```python
org_data.isna().sum()
```

```python
data = org_data
```

```python
# taking all numerical values
ndata = data.select_dtypes(include = np.number)
ndata.head()
```

```python
# getting range
ndata.max() - ndata.min()
```

```python
# first Quartile
Q1 = ndata.quantile(0.25)
Q1
```

```python
# second Quartile
Q2 = ndata.quantile(0.50)
Q2
```

```python
# third Quartile
Q3 = ndata.quantile(0.75)
Q3
```

```python
# Inter Quartile Range
IQR = Q3 - Q1
IQR
```

```python
ndata.shape
```

```python
# removing outliers
final_data = ndata[~((ndata < (Q1 - 1.5 * IQR)) |(ndata > (Q3 + 1.5 * IQR))).any(axis=1)]
final_data.shape
```

```python
# Correlation Matrix
plt.figure(figsize = (20,20))
corr = data.corr()
sns.heatmap(corr, annot=True);
plt.title("Correlation Matrix");
plt.show()
```

```python
sns.pairplot(data,hue='Output')
```

```python
# histogram of data
data.hist(figsize=(20,20));
```

# Machine learning

```python
# site and pop are correlated features
X = final_data.drop(['Output'], axis=1)
y = final_data['Output']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.kernel_ridge import KernelRidge
from xgboost.sklearn import XGBRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import SGDRegressor
```

```python
model1 = LinearRegression()
model2 = KernelRidge()
model3 = XGBRegressor()
```

```python
model4 = BayesianRidge()
model5 = ElasticNet()
model6 = SGDRegressor()
```

```python
model1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
# precting on training
pred = model1.predict(X_train)
rmse = (np.sqrt(mean_squared_error(y_train,pred)))
r2 = r2_score(y_train, pred)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# predicting on test
pred2 = model1.predict(X_test)
rmse = (np.sqrt(mean_squared_error(y_test, pred2)))
r2 = r2_score(y_test, pred2)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
